```python
sex=pd.get_dummies(tit_data['Sex'],drop_first=True)
sex.head()
```

```python
Pcl=pd.get_dummies(tit_data['Pclass'],drop_first=True)
Pcl.head()

```

```python
tit_data=pd.concat([tit_data,sex,Pcl],axis=1)
tit_data.head()
```

```python
tit_data.drop(['Pclass','Name','Sex'],axis=1,inplace=True)
```

```python
tit_data.head()
```

##Train Data

## Train Data

```python
X= tit_data.drop('Survived',axis=1)
y= tit_data['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
preds=log.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix as cm, accuracy_score as asc
classification_report(y_test,preds)
```

```python
cm(y_test,preds)
```

```python
asc(y_test,preds)*100
```
